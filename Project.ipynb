## $\textbf{Question}$: 

Given the observed range of masses $[m_{min},m_{max}]$ measured by 3dHST at its highest redshift:

1) Does the galaxiesâ€™ mass function ($\mathrm{GMF}$) in  $[m_{min},m_{max}]$ change as a function of redshift?
2) If so, which processes determine its evolution? It doesn't change, so we moved onto more interesting topics.

3) Found $\mathrm{GMF(z)}$, Can we predict how the mass below M_min is distributed at higher redshift? (apparently non-contemplabile, so ciaone).

$\textbf{Real question:}$
4) Given the SFR vs redshift projection of the 3-dimensional dataset (redshift, masses, sfr), are there some relations we can observe? Is there any functional dependency between the SFR and the $z$ that can be related to some model we developed? 

# 1) $\textbf{Closed box model}$:

## $\textbf{Assumptions for this model}$: 

- proportionality to the gas mass of the SFR
- close box assumption (only change in $m_{gas}$ is given by the SFR and not incoming external gas or gas leaving the galaxy $\rightarrow$ $m^{in}_{gas}=0 \ \ \ m^{out}_{gas}=0 $)
- $\epsilon$ is a constant 

![](Close_box_model.png)

We assume that the law describing the $\mathrm{SFR}$ as a function of time can be written as: 

$$\mathrm{SFR} = \epsilon m_{gas}(t)^{\alpha}$$

$$\frac{dm_{gas}(t)}{dt} = \epsilon m_{gas}(t)$$

Solving the equations we get: 

$$m_{gas} = m_{gas}(0) \cdot (1 - e^{-\epsilon (t-t_0)})$$

$$\mathrm{SFR} = \epsilon m_{gas, 0} \cdot e^{- \epsilon (t-t_0)}$$

$$\mathrm{sSFR} = \epsilon \frac{e^{-\epsilon(t-t_0)}}{1 - e^{-\epsilon(t-t_0)}}$$


## $\textbf{Discussion (points we stressed)}$:

- the $z$-taken data are strongly biased at higher $z$ values because of their relative distance from us, making it difficult to see smaller and less luminous masses. This means that our analysis can be considered more reliable if we choose only a specific range of masses to do statistics with. 

- the area increases at higher redshifts, and this leads to us seeing more rare elements ("big-L" galaxies, for instance) than the patch at smaller redshift (even if we learned that from 0.5 to 2 the differences are negligible, WTF)


# 2) $\textbf{Open box model}$:

## $\textbf{Assumptions for this model}$: 

- proportionality to the gas mass of the SFR
- $\rho_{DM}$ is a constant 
- Virialized mass = DM mass

![](Open_box_model.png)

We decided to adopt an accretion disk model around the central supermassive black hole of the galaxy. There will be an incoming mass as well as an outcoming mass that will respectively add up to the already present gas, while there will be processes contributing to the outcoming.

Let's restart from the equation above, by adding up two terms of incoming mass, and outcoming mass: 

$$\frac{dm_{gas}(t)}{dt} = \epsilon m_{gas}(t) + \dot{m}_{in} - \dot{m}_{out}$$

### $$\dot{m}_{out}$$

There may be a lot of reasons for the mass to get to the outside. Here we report some of the hypotheses on the processes: 

- AGM;
- accretion rate of the disk around the black hole (where we introduce a dependency on the $m_{bh}$; 
- supernovae ejection; 
- stripping of gas from another galaxy, the run pressure stripping process, which is only relevant at higher masses (e.g. very steep potential well). 

We concentrated on the supernovae ejections, basing our analysis on a very specific type of supernova explosion, the $\textit{type 1-A}$ kind. This was because we wanted to write the rate of ejection in terms of the $\texttt{SFR}$ such that: 

$$\dot{m}_{out} = \eta \texttt{SFR}$$

The equation above then becomes: 

$$\frac{dm_{gas}}{dt} = -(1 + \eta) \texttt{SFR} + \dot{m}_{in}$$

### $$\dot{m}_{in}$$

Here again we proposed some hypotheses: 

- The accretion of dark matter coming into the virialized sphere; 
- Galaxies mergers;

We went for the first option: 

From the above figure $dh$ is the thickness of the sphere, so that $dm = \rho dV$, where $\rho = \rho_{DM} = K$. Dividing the contribution from the 2D surface we observe: 

$$dV = \Sigma_{vir} dh = 4 \pi R_{vir}^2 dr$$

where $R_{vir}$ is the radius of the virialized part of the system. 

$\dot{m}_{vir}$ becomes: 


$$\dot{m}_{vir} = \rho_{DM} \Sigma_{DM} \frac{dh}{dt} = \rho_{DM} \Sigma_{DM} v_{DM} = \delta\rho_{0} (1+z)^3 = \delta\rho_{0} 4 \pi R_{vir}^2 \sqrt{\frac{G m_{vir}}{R_{vir}}} = ... = (\delta\rho_{0}G)^{1/2} (1+z)^{3/2} m_{vir}$$

The previous equation was: 

$$\frac{dm_{gas}}{dt} = -(1 + \eta) \texttt{SFR} + \dot{m}_{in}$$

We here have two options to solve it. 

### Option A

Consider the gas reservoir (our galaxy) at equilibrium. This means that $\frac{dm_{gas}}{dt} = 0$. 

From the graphs we see that the $\texttt{sSFR = \frac{\texttt{SFR}}{m_{*}}}$ is constant in mass; our goal is to check if this model can be compatible with this result. In the case of equilibrium we have this relation: 

$$\dot{m}_{gas} \propto \dot{m}_{DM} \propto \xi \dot{m}_{vir, in} \propto \xi m_{vir} (1+z)^{3/2}$$

where $\dot{M}_{DM} = $ is the accretion rate of dark matter evaluated above. ### A note on the $\epsilon$ value in $\texttt{SFR} = \epsilon m_{gas}$:

Considering the incoming baryonic matter: it needs to cool down to fall in. Why? Because otherwise the pressure due to the heating keeps on expanding the gas and there is no efficient accretion in the central bulge.
There are mass ranges in which it cools efficiently: 

$$\epsilon = \begin{cases} 0 \ \ \text{if} \ \ m_{vir} > 10^{12} M_{\odot} \\ 0 \ \ \text{if} \ \ m_{vir} < 10^{9} M_{\odot} \\ 1 \ \ \text{elsewhere} \end{cases}$$

There are then ranges for which the incoming mass is not enough to accrete around the bulge, so perhaps we need another process to probe. AGN?

```python
# mask each bin of redshift for the analysis:

def bin_redshift(matrix,step):
    i=np.min(matrix[:,0])
    masks=[]
    while(i<np.max(matrix[:,0])-step):
        masks.append(((matrix[:,0]>i) & (matrix[:,0]<i+step)))
        i=i+step
    return(masks)

# find optimal descriptive kde:

def kde_sklearn(data, bandwidth = 1.0):
    xgrid=np.linspace(np.min(data),np.max(data),1000)
    kde_skl = KernelDensity(bandwidth = bandwidth)
    kde_skl.fit(data[:, np.newaxis])
    log_pdf = kde_skl.score_samples(xgrid[:, np.newaxis]) # sklearn returns log(density)
    return np.exp(log_pdf)

# select best binning via CV and evaluate kde:

def KDE_binning(x):
    bwrange = np.linspace(0.01,1.0, 30) # Test 30 bandwidths from 0.1 to 1.0
    K = 5 # Do 5-fold cross validation
    grid = GridSearchCV(KernelDensity(), {'bandwidth': bwrange}, cv= K) # Try each bandwidth with K-folds
    grid.fit(x[:,None]) #Fit the histogram data that we started the lecture with.
    h_opt = grid.best_params_['bandwidth']
    pdf=kde_sklearn(x,h_opt)
    return pdf

def KDE_binning_LONGDATASET(x):
    bwrange = np.linspace(0.1, 1.0, 5) # Test 30 bandwidths from 0.1 to 1.0
    K = 2 # Do 5-fold cross validation
    grid = GridSearchCV(KernelDensity(), {'bandwidth': bwrange}, cv= K) # Try each bandwidth with K-folds
    grid.fit(x[:,None]) #Fit the histogram data that we started the lecture with.
    h_opt = grid.best_params_['bandwidth']
    pdf=kde_sklearn(x,h_opt)
    return pdf
    
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt    
from scipy import stats

# pip install uncertainties, if needed
try:
    import uncertainties.unumpy as unp
    import uncertainties as unc

except:
    try:
        from pip import main as pipmain
    except:
        from pip._internal import main as pipmain
    pipmain(['install','uncertainties'])
    import uncertainties.unumpy as unp
    import uncertainties as unc

n=len(yfit)

def f(x, a, b):
    return a * x + b

popt, pcov = curve_fit(f, xfit, yfit, sigma=errfit)

# retrieve parameter values
a = popt[0]
b = popt[1]
print('Optimal Values')
print('a: ' + str(a))
print('b: ' + str(b))

# compute r^2
r2 = 1.0-(sum((yfit-f(xfit,a,b))**2)/((n-1.0)*np.var(yfit,ddof=1)))
print('R^2: ' + str(r2))
print(pcov)
# calculate parameter confidence interval
a,b = unc.correlated_values(popt, pcov)
print('Uncertainty')
print('a: ' + str(a))
print('b: ' + str(b))

# plot data
plt.scatter(xfit, yfit, s=3, label='Data')
plt.errorbar(xfit, yfit, errfit, fmt='.')

# calculate regression confidence interval
px = np.linspace(0.04,0.5,100)
py = a*px+b
nom = unp.nominal_values(py)
std = unp.std_devs(py)

def predband(x, xd, yd, p, func, conf=0.95):
    # x = requested points
    # xd = x data
    # yd = y data
    # p = parameters
    # func = function name
    alpha = 1.0 - conf    # significance
    N = xd.size          # data sample size
    var_n = len(p)  # number of parameters
    # Quantile of Student's t distribution for p=(1-alpha/2)
    q = stats.t.ppf(1.0 - alpha / 2.0, N - var_n)
    # Stdev of an individual measurement
    se = np.sqrt(1. / (N - var_n) * \
                 np.sum((yd - func(xd, *p)) ** 2))
    # Auxiliary definitions
    sx = (x - xd.mean()) ** 2
    sxd = np.sum((xd - xd.mean()) ** 2)
    # Predicted values (best-fit model)
    yp = func(x, *p)
    # Prediction band
    dy = q * se * np.sqrt(1.0+ (1.0/N) + (sx/sxd))
    # Upper & lower prediction bands.
    lpb, upb = yp - dy, yp + dy
    return lpb, upb

lpb, upb = predband(px, xfit, yfit, popt, f, conf=0.95)

# plot the regression
plt.plot(px, nom, c='black', label='Linear Fit')

# uncertainty lines (95% confidence)
#plt.plot(px, nom - 1.96 * std, c='orange',\
#         label='95% Confidence Region')
#plt.plot(px, nom + 1.96 * std, c='orange')

# prediction band (95% confidence)
plt.plot(px, lpb, 'k--',label='Prediction Band')
plt.plot(px, upb, 'k--')
plt.ylabel('$\log(sSFR/yr^{-1})$')
plt.xlabel('log(1+z)')
plt.legend(loc='best')

# save and show figure
plt.savefig('regression.pdf')
plt.show()
```
